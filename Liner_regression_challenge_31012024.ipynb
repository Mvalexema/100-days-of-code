```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

```python
ins_data = pd.read_csv('insurance.csv')
ins_data.head()
```

```python
ins_data.shape
```

```python
x=ins_data.iloc[:,:1].values
y=ins_data.iloc[:,6:7].values
```

```python
print(x)
```

```python
print(y)
```

```python
ins_data.columns
```

```python
plt.scatter(x,y,color='b')
plt.xlabel('Age')
plt.ylabel('Expense')
plt.show()
```

```python
expense_model = LinearRegression()
expense_model.fit(x,y)

y_pred = expense_model.predict(x)

```

```python
plt.scatter(x,y,color = 'b')
plt.plot(x,y_pred, color = 'r')
plt.ylabel('Expense')
plt.xlabel('Age')
plt.show()
```
