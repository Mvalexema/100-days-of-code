```python

# Import libraries
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns
```

```python
data=pd.read_csv('Mall_Customers.csv')
data.head()
```

```python
data.describe()
```

```python


X = data.iloc[:,[3,4]].values

```

```python
import scipy.cluster.hierarchy as sch 
dendrogram=sch.dendrogram(sch.linkage(X,method='ward'))
plt.title('Dendrogram')
plt.xlabel('Customers')
plt.ylabel('Euclidean distances')
plt.show()
```

```python
sns.color_palette("tab10")
To_Plot=["Age","Genre","Annual Income (k$)","Spending Score (1-100)"]
print("Relative plot of selected features: a data subset")
plt.figure()
sns.pairplot(data[To_Plot], hue="Genre", palette=("tab10"))

plt.show()
```

```python
# Apply K-means clustering
kmeans = KMeans(n_clusters=4)  # Specify the number of clusters
kmeans.fit(X)
y_kmeans = kmeans.predict(X)

# Plotting the clusters
plt.scatter(X[:, 0], X[:, 1], c=y_kmeans, s=50, cmap='viridis')

centers = kmeans.cluster_centers_
plt.scatter(centers[:, 0], centers[:, 1], c='red', s=200, alpha=0.5)  # Plot the centroids
plt.title("K-means Clustering")
plt.xlabel("Income")
plt.ylabel("Score")
plt.show()
```
