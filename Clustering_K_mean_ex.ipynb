```python
# The K-means clustering algorithm is a method used to partition a dataset into K distinct, non-overlapping subgroups, where each data point belongs to the cluster with the nearest mean. 
# The process aims to minimize the variance within each cluster.
# This code does the following:
# It first generates a synthetic dataset using make_blobs, which creates blobs of points centered at random locations. 
# This dataset serves as an example.
# It then applies the K-means algorithm to this dataset by specifying the number of clusters (n_clusters=4 in this case).
# After fitting K-means to the data, it predicts the cluster for each data point, which is used to color the points in the plot.
# Finally, it plots the data points, coloring them based on their cluster, and also plots the centroids of these clusters in red.
```

```python
# Import necessary libraries
from sklearn.cluster import KMeans
from sklearn.datasets import make_blobs
import matplotlib.pyplot as plt

# Generate a sample dataset
X, _ = make_blobs(n_samples=300, centers=4, cluster_std=0.60, random_state=0)

# Apply K-means clustering
kmeans = KMeans(n_clusters=4)  # Specify the number of clusters
kmeans.fit(X)
y_kmeans = kmeans.predict(X)

# Plotting the clusters
plt.scatter(X[:, 0], X[:, 1], c=y_kmeans, s=50, cmap='viridis')

centers = kmeans.cluster_centers_
plt.scatter(centers[:, 0], centers[:, 1], c='red', s=200, alpha=0.5)  # Plot the centroids
plt.title("K-means Clustering")
plt.xlabel("Feature 1")
plt.ylabel("Feature 2")
plt.show()
```
