```python
# Import libraries
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns
```

```python
data=pd.read_csv('data_freedom.csv')
data.head()
```

```python
data.describe()
```

```python
X = data.iloc[:,[4,5]].values
```

```python
sns.color_palette("tab10")
To_Plot=["Overall Score","Property Rights","Government Integrity","Judicial Effectiveness","Tax Burden"]
print("Relative plot of selected features: a data subset")
plt.figure()
sns.pairplot(data[To_Plot])

plt.show()
```

```python
# Apply K-means clustering
kmeans = KMeans(n_clusters=4)  # Specify the number of clusters
kmeans.fit(X)
y_kmeans = kmeans.predict(X)

# Plotting the clusters
plt.scatter(X[:, 0], X[:, 1], c=y_kmeans, s=50, cmap='viridis')

centers = kmeans.cluster_centers_
plt.scatter(centers[:, 0], centers[:, 1], c='red', s=200, alpha=0.5)  # Plot the centroids
plt.title("K-means Clustering")
plt.xlabel("Overall Score")
plt.ylabel("Property Rights")
plt.show()
```
