```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression


data = pd.read_csv("diabetes.csv", index_col = 0)
data.head()
```

```python
#Simple linear regression

x = data.iloc[:,0].values
print(x)
y = data.iloc[:,12].values
y=y.reshape(-1,1)
x=x.reshape(-1,1)
```

```python
diab_model = LinearRegression()

diab_model.fit(x,y)
y_pred = diab_model.predict(x)
plt.scatter(x,y,color = 'b')
plt.plot(x,diab_model.predict(x),color = 'r')
plt.title('\encounter_id vs num_lab_procedures')
plt.xlabel('num_lab_procedures')
plt.ylabel('encounter_id')
plt.show()


```
