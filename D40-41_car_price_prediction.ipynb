```python
import numpy as np 
import pandas as np 
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import LabelEncoder 
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error, r2_score

```

```python
df=np.read_csv('CarPrice_Assignment.csv')
```

```python
X=df.loc[::,'horsepower'].values
y=df.loc[::,'price'].values

plt.scatter(X,y,color='r')
```

```python


X = X.reshape(-1,1)
y=y.reshape(-1,1)

X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.25)


X_test.shape

model = LinearRegression()
model.fit(X_train, y_train)
```

```python
y_pred = model.predict(X_test)

plt.scatter(X_test,y_test,color='b')
plt.plot(X_test,y_pred,color='m')

plt.xlabel('Enginesize')
plt.ylabel('Price')
plt.show()
```

```python
print(f"The price of a vehicle with the 150 horsepower: {model.predict([[150]])}")
```

```python
score = model.score(X_test, y_test)
score
```

```python
le=LabelEncoder()
var_mod=df.select_dtypes(include='object').columns

for i in var_mod:
    df[i] = le.fit_transform(df[i])
```

```python
X=df.drop(['price'],axis=1)
y=df['price']
X_train, X_test,y_train,y_test = train_test_split(X,y,test_size=0.25)
```

```python
model=LinearRegression()
model.fit(X_train,y_train)
```

```python
y_pred = model.predict(X_test)
```

```python
mse=mean_squared_error(y_test,y_pred)
r2=r2_score(y_test,y_pred)
```

```python
print("Mean Square Error:", mse)
print("R-squared:", r2)
```
